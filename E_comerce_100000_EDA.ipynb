```python

print('')
print('='*60)
print('Parsing Date column')
print('='*60)
print('')
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['OrderDate'].dt.month.value_counts().sort_index()


print('')
print('='*60)
print('Monthly production')
print('='*60)
print('')
print(f'The Monthly production : \n {df.groupby(df['OrderDate'].dt.month)['TotalAmount'].sum()}')

print('')
print('='*60)
print('Yearly production')
print('='*60)
print('')


print(f'The Yearly production : \n {df.groupby(df['OrderDate'].dt.year)['TotalAmount'].sum()}')
```
